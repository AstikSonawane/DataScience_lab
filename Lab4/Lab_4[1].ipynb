## Hypothesis Testing

```python
import pandas as pd
from scipy import stats

# Read the Excel file with correct parameters
data = pd.read_excel(r'D:\SEM6\DSA Lab\Lab 4_DSA.xlsx', header=3)#Skip the first three rows as they are
#  empty or have irrelevantinformation

# Extract mean call duration for Day 10
mean_duration_day_10 = data.loc[data['Day'] == 10, 'Mean Call Duration'].values[0]

# Assuming the population mean (Î¼) is known or given
population_mean = 4  # Assuming the population mean is 8.5 (you can change it accordingly)

# Calculate the sample standard deviation
sample_stddev = 3

# Assuming the sample size (n)
sample_size = 50

# Calculate the standard error (SE)
standard_error = sample_stddev / (sample_size**0.5)

# Calculate the t-value using the formula: t = (x_bar - mu) / (s / sqrt(n))
t_value = (mean_duration_day_10 - population_mean) / standard_error

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the p-value using the cumulative distribution function (CDF) of the t-distribution
p_value = stats.t.cdf(t_value, df=degrees_of_freedom) * 2  # Multiply by 2 for a two-tailed test

print("Mean Call Duration for Day 10:", mean_duration_day_10)
print("Sample Standard Deviation:", sample_stddev)
print("T-value:", t_value)
print("P-value:", p_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of t-values for the t-distribution
t_values = np.linspace(-4, 4, 1000)

# Calculate the probability density function (PDF) of the t-distribution
pdf = stats.t.pdf(t_values, df=degrees_of_freedom)

# Plot the t-distribution
plt.plot(t_values, pdf, label='t-distribution')

# Highlight the critical t-values for a two-tailed test at alpha = 0.05
critical_t_values = [stats.t.ppf(0.025, df=degrees_of_freedom), stats.t.ppf(0.975, df=degrees_of_freedom)]
plt.axvline(critical_t_values[0], color='r', linestyle='--', label='Critical t-value (alpha=0.05)')
plt.axvline(critical_t_values[1], color='r', linestyle='--')

# Mark the calculated t-value on the graph
plt.axvline(t_value, color='g', linestyle='-', label='Calculated t-value')

# Set labels and title
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.title('t-Distribution with Critical Values and Calculated t-value')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of t-values for the t-distribution
t_values = np.linspace(-4, 4, 1000)

# Calculate the probability density function (PDF) of the t-distribution
pdf = stats.t.pdf(t_values, df=degrees_of_freedom)

# Plot the t-distribution
plt.plot(t_values, pdf, label='t-distribution')

# Highlight the critical t-values for a two-tailed test at alpha = 0.05
critical_t_values = [stats.t.ppf(0.025, df=degrees_of_freedom), stats.t.ppf(0.975, df=degrees_of_freedom)]
plt.axvline(critical_t_values[0], color='r', linestyle='--', label='Critical t-value (alpha=0.05)')
plt.axvline(critical_t_values[1], color='r', linestyle='--')

# Mark the calculated t-value on the graph
plt.axvline(t_value, color='g', linestyle='-', label='Calculated t-value')

# Shade the region under the t-distribution curve corresponding to the p-value
if p_value < 0.05:
    plt.fill_between(t_values[t_values <= t_value], 0, pdf[t_values <= t_value], color='green', alpha=0.3, label='Area corresponding to P-value')
else:
    plt.fill_between(t_values[t_values >= t_value], 0, pdf[t_values >= t_value], color='green', alpha=0.3, label='Area corresponding to P-value')

# Set labels and title
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.title('t-Distribution with Critical Values and Calculated t-value')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import stats

# Salmonella levels in 9 randomly sampled batches of ice cream
salmonella_levels = np.array([0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418])

# Define the null hypothesis: mean level of salmonella in ice cream is not greater than 0.3
null_hypothesis_mean = 0.3

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(salmonella_levels, null_hypothesis_mean, alternative='greater')

# Print the t-statistic and p-value
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Check if the p-value is less than the significance level (e.g., 0.05)
significance_level = 0.05
if p_value < significance_level:
    print("There is evidence that the mean level of salmonella in ice cream is greater than 0.3.")
else:
    print("There is no evidence that the mean level of salmonella in ice cream is greater than 0.3.")
```

```python
# A drug(treatement group) an additional 6 subjects a plassiobo (control grp) there reaction time to a stimulus was measured (milisec) we want to perform a 2 sample t test for comapring the means of treatment and control groups 
# control 91 87 99 77 88 91
# treat  101 110 103 93 99 104
# is drug is equivalent to plassibo or not

import numpy as np
from scipy import stats

# Reaction times for the control group
control = np.array([91, 87, 99, 77, 88, 91])

# Reaction times for the treatment group
treatment = np.array([101, 110, 103, 93, 99, 104])

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(treatment, control)

# Print the t-statistic and p-value
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Set the significance level
significance_level = 0.05

# Compare the p-value with the significance level
if p_value < significance_level:
    print("The drug has a statistically significant effect compared to the placebo.")
else:
    print("There is no statistically significant difference between the drug and the placebo.")
```
